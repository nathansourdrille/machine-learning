# Optmisation sans contraintes 

L'objectif est d'utiliser le package Python CVXPY pour résoudre des problèmes d'optimisation convexe allant d'un exemple introductif à des problèmes plus élaborés.

## A simple example 

Résolution sur papier 

```python
import numpy as np
import cvxpy as cvx
from utility import plot_contours_exercice_section2
import matplotlib.pyplot as plt
```

```python
c = np.array([3, 1])
# contraintes linéaires 
b = np.ones(4)
A = np.array([[1, 1], [1, -1], [-1, 1], [-1, -1]])
```

```python
plot_contours_exercice_section2(A, b, c)
```

**courbes de niveau**: ensemble de points où la fonciton prend la même valeur.

Le carré bleu en (3,1) correspond au vecteur c.  
Le carré noir représente l'ensemble des points où les contraintes sont respectées.  
Le point le plus proche de c en respectant les contraintes se trouve donc en (1,0).

### Problème primal

```python
# définit la dimension de la variable theta
d = 2
# créé un vecteur de dimension d manipulable par cvx
theta = cvx.Variable(d)
# fonction objectif à minimiser
obj = cvx.Minimize(0.5*cvx.quad_form(theta-c, np.eye(d)))
# défini les contraintes linéaires (Aθ ≤ b)
constraints = [A@ theta - b <= 0]
# préparation de la résolution du problème primal
primal = cvx.Problem(obj, constraints)
# résolution du problème primal
primal.solve(verbose = False)

print(f"état de la solution = {primal.status}")
print(f"Solution primal optimale = {theta.value}")
obj_primal = 0.5*cvx.quad_form(theta-c, np.eye(d))
print(f"Fonction objectif primale à l'optimalité = {obj_primal.value}")
```

La solution obtenue correspond à celle devinée intuitivement à partir du graphe (1,0).

### Problème dual

```python
H = A@A.T
q = -(A@c - b)
# nombre de contraintes
m = 4 
mu = cvx.Variable(m) 

obj = cvx.Minimize(0.5*(cvx.quad_form(mu, H))+mu.T@q)

constraints = [mu >= 0]

dual = cvx.Problem(obj, constraints)
dual.solve(verbose = False) 

theta_from_dual = c - np.asarray(A.T@(mu.value))

print(theta_from_dual)
```

On trouve la même solution mais avec plus de précision pour le problème dual (valeurs exactes).

## Minimum enclosing ball

Les variables inconnues sont R et z. Il y a une contrainte par maison.

$$
\begin{align*}
\mathcal{L}(R, z, \mu) = R^2 + \sum_{i=1}^n \mu_i (\lVert z - x_i \rVert_2^2 - R^2)
\end{align*}
$$

La condition de stationnarité par rapport à R donne : 
$$
\begin{align*}
& \nabla_R \mathcal{L}(R, z, \mu) = 0 \\
\implies & 2R - 2 R \sum_{i=1}^n \mu_i = 0 \\
\implies & 2R \left( 1 - \sum_{i=1}^n \mu_i \right) = 0 \\
\end{align*}
$$

La solution R = 0 est impossible donc la condition de stationnarité par rapport à R donne :

-$1 - \sum_{i=1}^n \mu_i = 0 \implies \sum_{i=1}^n \mu_i = 1$

Celle par rapport à z : 

$$
\begin{align*}
& \nabla_z \mathcal{L}(R, z, \mu) = 0 \\
\implies & \sum_{i=1}^n \mu_i \nabla_z \lVert z - x_i \rVert_2^2 = 0 \\
\implies & \sum_{i=1}^n \mu_i \frac{z - x_i}{\lVert z - x_i \rVert_2^2} = 0 \\
\implies & z = \sum_{i=1}^n \mu_i x_i
\end{align*}
$$

En remplaçant les 2 équations précédemment trouvées dans notre lagrangien et en développant la norme 2 au carré de la somme des mu i xi - xi on obtient : 

$$
\begin{align*}
\mathcal{L}(R, z, \mu) = -\sum_{i=1}^n \sum_{j=1}^n \mu_i \mu_j x_j^\top x_i + \sum_{i=1}^n \mu_i x_i^\top x_i \\
\end{align*}
$$

$$
\begin{align*}
\max_{\mu} & -\sum_{i=1}^n \sum_{j=1}^n \mu_i \mu_j x_j^\top x_i + \sum_{i=1}^n \mu_i x_i^\top x_i \\
\text{sous contraintes} & 
\begin{cases}
\mu_i \geqslant 0, \forall i \in 1, \ldots, n \\
\sum_{i=1}^n \mu_i = 1
\end{cases}
\end{align*}
$$

On calcule z grâce à cette équation précedemment trouvée : 

$$
z = \sum_{i=1}^n \mu_i x_i
$$

et on calcule R avec : 

$$
R = \max_{i}{\lVert z - x_i \rVert_2^2}
$$

```python
X = np.genfromtxt("maisons.csv", delimiter=",")
print(X)
```

But: trouver la boule la plus petite qui englobe les points et identifier les points critiques qui déterminent sa taille.

```python
# Matrice et vecteurs du problème dual
H = X.dot(X.T)        # H = X * X^T, matrice quadratique du dual
q = np.diag(H)        # q = diagonale de H
n = H.shape[0]        # dimension du vecteur dual
e = np.ones(n)        # vecteur de 1, utilisé pour la somme des mu
zer = np.zeros(n)     # vecteur de 0, utilisé pour la contrainte mu >= 0

# mu est défini comme la variable d'optimisation duale
mu = cvx.Variable(n)

# Définition de la fonction objectif du problème dual
obj = cvx.Minimize(cvx.quad_form(mu, H) - mu.T @ q)
# cvx.quad_form(mu, H) = mu^T H mu
# mu.T @ q = produit linéaire avec q
# On cherche à MINIMISER l'objectif dual

# Définition des contraintes du problème dual
constr = mu >= zer, cvx.sum(mu) == e
# mu >= 0 : les variables duales sont positives
# cvx.sum(mu) == e : la somme des mu doit être égale à e (vecteur de 1)

# Résolution du problème d'optimisation contraint
prob = cvx.Problem(obj, constr)
prob.solve(verbose=False)  # verbose=False supprime l'affichage du solveur

# Extraction de la solution duale sous forme de vecteur (ndarray)
mu = np.squeeze(np.asarray(mu.value))

# Affichage de la solution duale
print('Solution duale mu = ', mu)

# Affichage de la valeur de la fonction objectif duale
print("Valeur de la fonction objectif = ", -1.0 * prob.value)
```

```python
# Calcul du centre z
z = np.zeros(2)  # initialisation du centre à zéro
for i in range(n):
    z += mu[i] * X[i, :]  # sommation pondérée des vecteurs X par les coefficients duals mu

# Une manière matricielle de calculer z :
# z = np.multiply(X, np.outer(mu, np.ones(X.shape[1]))).sum(axis=0)
# np.outer(mu, np.ones(X.shape[1])) crée une matrice où chaque ligne i est mu[i] * 1
# np.multiply applique la multiplication élément par élément avec X
# sum(axis=0) fait la somme des lignes pour obtenir z

print("Le centre z = ", z)

# Calcul du rayon R
seuil = 1e-3
pos = np.where(mu >= seuil)  # indices des coefficients mu significatifs (non nuls)

# Calcul de la distance quadratique entre z et chaque point X[i,:]
Distancequad = z.dot(z) - 2*(X.dot(z)) + q  # correspond à ||z - X[i]||^2
# Ici q = diag(H) = ||X[i]||^2, donc Distancequad[i] = ||z||^2 - 2 <X[i], z> + ||X[i]||^2

R2 = np.mean(Distancequad[pos])  # moyenne des distances quadratiques pour les points actifs (mu >= seuil)
print("R^2 : ", R2)
```

```python
fig = plt.figure(figsize=(8,8))

# Tracé des points X (les "maisons") avec des carrés bleus
plt.plot(X[:,0], X[:,1], "s", color="b", markerfacecolor="b", markersize=10)

# Tracé du centre z avec un marker rouge en forme de pentagone
plt.plot(z[0], z[1], "rp", markersize=15)

# Tracé de la "boule minimale englobante" (minimum enclosing ball)
t = np.arange(0, 2*np.pi+0.02, 0.01)  # création d'un vecteur d'angles pour dessiner le cercle
plt.figure(fig.number)  # on dessine sur la même figure
plt.plot(z[0] + np.sqrt(R2)*np.cos(t), z[1] + np.sqrt(R2)*np.sin(t), "m")  
# "m" = magenta ; cercle de rayon sqrt(R2) centré en z

# Tracé des points "les plus éloignés" qui définissent la boule (ceux situés sur le cercle)
plt.plot(X[pos,0], X[pos,1], "og", alpha=0.5, markersize=15, linewidth=2)
# "og" = cercle vert ; alpha=0.5 pour semi-transparence

plt.show()
```

**outlier**: point qui s'écarte fortement de la tendance générale des données. utiles pour:
- les exclure si nécessaire.
- détecter des erreurs de mesure.
- analyser les évènement rares.

```python
outliers = [
[3.0000, -3.0000],
[2.5000, 2.5000]
]
Xoutliers = np.array(outliers)
```

```python
fig = plt.figure(figsize=(8,8))
plt.plot(X[:,0], X[:,1], "s", color="b", markerfacecolor="b", markersize = 10)
plt.plot(z[0], z[1], "rp", markersize=15)

t = np.arange(0, 2*np.pi+0.02, 0.01)
plt.figure(fig.number)
plt.plot(z[0] + np.sqrt(R2)*np.cos(t), z[1] + np.sqrt(R2)*np.sin(t), "m")

plt.plot(X[pos,0], X[pos,1], "og", alpha=0.5, markersize = 15, linewidth =2)
plt.plot(Xoutliers[:,0], Xoutliers[:,1], "o", label="outlier", color="orange")
plt.legend()
plt.show()
```

Le problème est le même en augmentant les dimensions de 2 aux endroits nécessaires.

```python
X_o = np.concatenate((X, Xoutliers))
print(X_o)
```

```python
H_o = X_o.dot(X_o.T)
q_o = np.diag(H_o)
n_o = H_o.shape[0]
e_o = np.ones(n_o)
zer_o = np.zeros(n_o)
mu_o = cvx.Variable(n_o)
obj_o = cvx.Minimize(cvx.quad_form(mu_o, H_o) - mu_o.T@q_o)
constr_o = [mu_o >= zer_o, cvx.sum(mu_o)==e_o]
prob = cvx.Problem(obj_o, constr_o)
prob.solve(verbose = False)
mu_o = np.squeeze(np.asarray(mu_o.value))
print('Dual solution mu = ', mu_o)
print("Objective function value = ", -1.0*prob.value)
```

```python
z_o = np.zeros(2)
for i in range(n_o):
    z_o += mu_o[i]*X_o[i,:]
print("The center z = ", z_o)
seuil_o = 1e-3
pos_o = np.where(mu_o >= seuil)
Distancequad_o = z_o.dot(z_o) - 2*(X_o.dot(z_o)) + q_o
R2_o = np.mean(Distancequad_o[pos_o])
print("R^2 : ", R2_o)
```

```python
fig_o = plt.figure(figsize=(8,8))
plt.plot(X_o[:,0], X_o[:,1], "s", color="b", markerfacecolor="b", markersize = 10)
plt.plot(z_o[0], z_o[1], "rp", markersize=15)
t_o = np.arange(0, 2*np.pi+0.02, 0.01)
plt.figure(fig_o.number)
plt.plot(z_o[0] + np.sqrt(R2_o)*np.cos(t_o), z_o[1] + np.sqrt(R2_o)*np.sin(t_o), "m")
plt.plot(X_o[pos_o,0], X_o[pos_o,1], "og", alpha=0.5, markersize = 15, linewidth =2)
plt.show()
```

On remarque ici que le cercle a été agrandi en conséquence pour inclure les maisons les plus éloignées.

Mais en faisant cela la caserne de pompiers a été éloignée des gros pâtés de maisons. En ce sens là nous avons rendu la solution moins optimale.

Pour régler ceci nous pouvons imaginer poser un nouveau problème où les maisons ont toutes un poinds et une maison proches d'autres maisons a un poids plus important comme ça nous privilégions les ensembles de maisons proches au lieu de prendre chaque maison de manière unique sans considérer ses voisins

En conclusion, dans une première partie, nous avons formulé un problème quadratique sous contraintes linéaires, établi son Lagrangien et les conditions KKT, avant d’en déduire le problème dual. La comparaison entre les solutions du primal et du dual, obtenues à l’aide de cvxpy, a confirmé la cohérence des deux approches et illustré le principe de dualité forte en optimisation convexe.  
Dans une seconde partie, nous avons étudié le problème de la sphère englobante minimale, appliqué ici à la localisation optimale d’une caserne de pompiers. Cette modélisation a montré comment les contraintes influencent la solution, et comment le dual permet de retrouver le centre et le rayon optimaux. L’ajout de points aberrants a également permis d’observer la sensibilité du modèle et d’ouvrir la réflexion sur des approches plus robustes.

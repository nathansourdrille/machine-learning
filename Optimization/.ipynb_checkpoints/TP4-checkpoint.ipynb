# Optimisation sans contrainte 

L'objectif est d'implémenter une descente de gradient afin de résoudre un problème d'optimisation sans contraintes. On s'interessera aussi à la méthode de Newton.

## Unconstrainted Problem

Nous considérons le problème de minimisation suivant :

$$
\min_{\theta \in \mathbb{R}^2} J(\theta)
$$

où

$$
J(\theta) = (1 - \theta_1)^2 + 100\,(\theta_2 - \theta_1^2)^2.
$$

Il s'agit de la fonction de Rosenbrock.


## Our goal

Déterminons d'abord à la main le gradient de la fonction et son point de stationnarité. Ensuite, on va déterminer la Hessienne.

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.linalg import norm
img = cv2.imread('resolution.jpg', -1)
plt.figure(figsize=(25,25))
plt.imshow(np.rot90(img, k=3));
```

Le point stationnaire de la fonction est en (1,1).

```python
# Hessienne
def myhessian(theta):
    t1, t2 = float(theta[0]), float(theta[1])
    return np.array([[2+1200*t1**2 - 400*t2, -400*t1], [-400*t1, 200.0]])
```

```python
theta_etoile = np.array([1.0, 1.0])
# calcule de la Hessienne en (1,1)
H_etoile = myhessian(theta_etoile)
print(f"H(θ*)= {H_etoile}")
```

```python
# la Hessienne est une matrice carré, elle est def pos si ses valeurs propres sont positives
val_propres = np.linalg.eigvalsh(H_etoile)
print(f"Valeurs propres de H(θ*): {val_propres}")
```

**Les valeurs propres sont strictement positives donc la Hessienne au point (1,1) est définie positive. Donc le point stationnaire θ\* est un minimum local de J.**

## ... and how we reach it

Le **gradient** est un vecteur qui caractérise la variation d'une fonction à plusieurs variables. Il correspond au vecteur des dérivées partielles d'une fonction.  
Pour savoir dans quelle direction f augmente le plus rapidement, on considère une direction u (vecteur unitaire) et la dérivée directionnelle. On obtient un produit scalaire entre le gradient de f et la direction u, ce produit scalaire est maximumal lorsque u est aligné avec le gradient. 

Dans la descente de gradient standard, la direction de descente correspond à l'opposé du gradient:  **hk​ = −∇J(θk​)**.  
En effet, **le gradient indique la direction de plus forte augmentation locale de la fonction**, or on cherche à diminuer cette fonction.

Aussi, la norme du gradient correspond au taux de variaton maximal:
- gradient grand => f augmente très vite dans cette direction.
- gradient petit => f augmente lentement dans cette direction.

Donc le gradient indique à la fois la direction de la pente la plus raide et la rapidité avec laquelle la fonction change.

```python
def mycriterion(theta):
    t1, t2 = theta[0], theta[1]
    J = (1-t1)**2 + 100*(t2-t1**2)**2
    return J
```

```python
def mygradient(theta):
    t1, t2 = theta[0], theta[1]
    gradJ = np.array([-2*(1-t1) - 400*t1*(t2-t1**2), 200*(t2-t1**2)])
    return gradJ
```

```python
# visualisation de la fonction de Rosenbrock
n = 100
points_x1, points_x2 = np.meshgrid(np.linspace(-1.25, 1.5, n), np.
linspace(-1.75, 1.5, n))
f = (1-points_x1)**2 + 100*((points_x2 - points_x1**2)**2)
f = f.reshape(points_x1.shape)
levels = np.concatenate((np.array([0, 1]), np.arange(5, 45, 5)))
fig = plt.figure(1, figsize=(8,4))
cp = plt.contourf(points_x1, points_x2, f, levels, alpha=0.95, cmap="RdBu")
plt.colorbar()

# point initial
theta0 = np.array([-1.0, 0.0])
plt.figure(fig.number)
plt.scatter(theta0[0], theta0[1], marker="o", color="k", facecolor="k", s
=150)
plt.text(-1.1, -0.5, r"${\theta}_0$", {"color": "k", "fontsize": 20})
plt.xticks(fontsize=16), plt.yticks(fontsize=16);
```

```python
# nombre maximal d'itérations
iter_max = 8000

# seuil 
thresh = 1e-2
alpha = 1e-3

theta = theta0.copy()
iter = 0

# stockage des résultats
history_J = np.empty(iter_max)
history_theta = np.empty((theta0.shape[0], iter_max)) 

history_theta[:,iter] = theta0
history_J[iter] = mycriterion(theta0)
grad = mygradient(theta0)

while (iter <= iter_max-2) and (np.linalg.norm(grad) > thresh):
    # calcule de la direction de descente
    direction = -grad
    # le pas
    # mise à jour de la solution
    theta += alpha*direction
    iter += 1
    history_theta[:,iter] = theta
    history_J[iter] = mycriterion(theta)
    grad = mygradient(theta)
print(f"Solution finale: {theta}")
iter_count = iter
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(iter_count), history_J[:iter_count], 'b-o', markersize=4)
plt.xlabel('Iteration', fontsize=14)
plt.ylabel('J(theta)', fontsize=14)
plt.title('Evolution de J selon les itérations', fontsize=16)
plt.grid(True)
plt.show()
```

```python
n = 100
points_x1, points_x2 = np.meshgrid(
    np.linspace(-1.25, 1.5, n),
    np.linspace(-1.75, 1.5, n)
)

f = (1 - points_x1)**2 + 100*((points_x2 - points_x1**2)**2)
f = f.reshape(points_x1.shape)

levels = np.concatenate((np.array([0, 1]), np.arange(5, 45, 5)))

fig = plt.figure(1, figsize=(8,4))
cp = plt.contourf(points_x1, points_x2, f, levels, alpha=0.95, cmap="RdBu")
plt.colorbar()

# point initial
theta0 = np.array([-1.0, 0.0])
plt.scatter(theta0[0], theta0[1], marker="o", color="k", facecolor="k", s=150)
plt.text(-1.1, -0.5, r"${\theta}_0$", {"color": "k", "fontsize": 20})

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# on récupère le code précédent pour afficher graphiquement la convergence vers la solution optimale

iter_count = np.where(history_J != 0)[0].max() + 1

# tracer la trajectoire
plt.plot(history_theta[0, :iter_count], history_theta[1, :iter_count],
         marker='o', color='yellow', markersize=5, linewidth=2, label='Positions successives de θ')

# solution finale
theta_hat = history_theta[:, iter_count-1]
plt.scatter(theta_hat[0], theta_hat[1], color='green', s=120, label=r'$\hat{\theta}$')

# minimum global
theta_star = np.array([1.0, 1.0])
plt.scatter(theta_star[0], theta_star[1], color='red', s=120, label=r'$\theta^*$')

plt.legend(fontsize=12)
plt.title('Descente de gradient de la fonction de Rosenbrock', fontsize=16)
plt.show()
```

On converge vers la solution optimale cependant on ne l'atteint pas, malgré 8k itérations. Dès lors, la qualité de résolution est bonne mais pas très satisfaisante. On va essayer une autre méthode: Newton.

## Newton method

```python
# implémentation de la méthode de Newton

# paramètres
iter_max = 10
thresh = 1e-6
lambda_reg = 1e-3 
theta0 = np.array([-1.0, 0.0])

history_J = []
history_theta = []

theta = theta0.copy()
grad = mygradient(theta)

history_theta.append(theta.copy())
history_J.append(mycriterion(theta))

iter = 0

# boucle Newton
while iter < iter_max-1 and norm(grad) > thresh:
    H = myhessian(theta)
    
    # régularisation pour que H soit définie positive
    H_reg = H + lambda_reg * np.eye(2)
    
    # direction de Newton : multiplication matrice-vecteur, négatif pour descente
    direction = -np.linalg.inv(H_reg) @ grad
    
    # mise à jour du point courant
    theta += direction
    
    # stockage
    history_theta.append(theta.copy())
    history_J.append(mycriterion(theta))
    
    # calcul du nouveau gradient
    grad = mygradient(theta)
    iter += 1


history_theta = np.array(history_theta).T  
history_J = np.array(history_J)            

theta_hat = theta
print(f"Solution finale (Newton) : {theta_hat}")
```

```python
n = 100
points_x1, points_x2 = np.meshgrid(
    np.linspace(-1.25, 1.5, n),
    np.linspace(-1.75, 1.5, n)
)

f = (1 - points_x1)**2 + 100*((points_x2 - points_x1**2)**2)
f = f.reshape(points_x1.shape)

levels = np.concatenate((np.array([0, 1]), np.arange(5, 45, 5)))

fig = plt.figure(1, figsize=(8,4))
cp = plt.contourf(points_x1, points_x2, f, levels, alpha=0.95, cmap="RdBu")
plt.colorbar()

# point initial
theta0 = np.array([-1.0, 0.0])
plt.scatter(theta0[0], theta0[1], marker="o", color="k", facecolor="k", s=150)
plt.text(-1.1, -0.5, r"${\theta}_0$", {"color": "k", "fontsize": 20})

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# on récupère le code précédent pour afficher graphiquement la convergence vers la solution optimale

iter_count = np.where(history_J != 0)[0].max() + 1

# tracer la trajectoire
plt.plot(history_theta[0, :iter_count], history_theta[1, :iter_count],
         marker='o', color='yellow', markersize=5, linewidth=2, label='Positions successives de θ')

# solution finale
theta_hat = history_theta[:, iter_count-1]
plt.scatter(theta_hat[0], theta_hat[1], color='green', s=120, label=r'$\hat{\theta}$')

# minimum global
theta_star = np.array([1.0, 1.0])
plt.scatter(theta_star[0], theta_star[1], color='red', s=120, label=r'$\theta^*$')

plt.legend(fontsize=12)
plt.title('Méthode de newton pour la fonction de Rosenbrock', fontsize=16)
plt.show()
```

On obtient la solution optimale en moins de 10 itérations, c'est bien plus rapide et précis que la descente de gradient classique.

En conclusion, pour Rosenbrock 2D, la méthode de Newton est beaucoup plus efficace que la descente de gradient. En effet, la descente de gradient nécessite beaucoup plus d'itérations et un réglage précis du pas. Comme on le voit graphiquement, Newton a des très grands sauts initiaux car la Hessienne est mal conditionnée, cependant elle converge très rapidement car dès la 4ème itération on est proche de la solution.

Alors, bien que la descente de gradient soit une méthode simple et plutôt stable, elle est inefficace pour les vallées étroites et courbées comme Rosenbrock 2D, ce qui amène à une convergence très lente.
Cependant, la méthode de Newton converge très rapidement et sa trajectoire est efficace mais elle nécessite des calculs coûteux avec la Hessienne, qui nécessite d'ailleurs une régularisation pour être bien définie positive.
